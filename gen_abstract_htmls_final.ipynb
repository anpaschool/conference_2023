```python
def single_function_html_v2(abstract_id, presentation_title, author_name, author_aff,
                          coauthor_names,coauthor_affiliations, abstract_text, Division, year):
    
    name_list = coauthor_names.split(';')
    name_list.insert(0,author_name)
    aff_list = coauthor_affiliations.split(';') 
    aff_list.insert(0,author_aff)
    max_len = max(len(name_list),len(aff_list))
    name_list_new = name_list +[" "]*(max_len -len(name_list))
    aff_list_new = aff_list +[" "]*(max_len -len(aff_list))

    str1 = f"""
    <body>
    <main class="container">
    <div class="row">
      <div class="col mt-5">
            <h2>Bulletin of ANPA </h2>
            <h3>Abstract submitted to ANPA Conference July 14&#8211;16, {year}</h3>
            <h3>Volume 5, Number 1</h3>
                <h5>{Division}</h5>
                <h6>Abstract ID: {abstract_id}</h6>
                <h4 class="mt-4">Abstract:</h4>
        <p>
                <span title="{abstract_id}">{abstract_id}: </span>
                {presentation_title}
                </p>
            <h4 class="mt-4">Authors:</h4>
    """
    str2 =[]
    for name,aff in zip(name_list_new,aff_list_new):
        if (name!=" ") or (aff!=" "):
            str2.append(f"""
        <ul>
            <li>{name}; {aff}</li>
        </ul>           
    """)
        
    str3 = f"""
    <p class="my-4">
            {abstract_text}
            </p>
    <p><small>To cite this abstract, use the following reference: https://anpaglobal.org/conference/{year}/{abstract_id}</small></p>
                </div>
            </div>     
        </main>
    </body>
    </html>"""

    str4 = f"""
    <!DOCTYPE html>
    <html lang="en">

    <head>
        <meta charset="UTF-8">
        <title>ANPA Conference {year} - {str(presentation_title)} </title>
        <meta name="citation_conference" content="Abstract submitted to ANPA Conference 2023">
        <meta name="citation_publisher" content="Association of Nepali Physicists in America">

        <meta name="citation_title" content={str(presentation_title)}>
        <meta name="citation_author" content={str(author_name)}>
    """
    str5 = []
    for name in name_list:
        str5.append(f""" 
        <meta name="citation_author" content={str(name)}>
        """)
        
    str6 = f"""
        <meta name="citation_publication_date" content="{year}/07/14">
        <meta name="citation_conference_title" content="Proceedings of ANPA Conference {year}">
        <meta name="citation_volume" content="{year}">
        <meta name="citation_issue" content="1">
        <meta name="citation_pdf_url" content="https://www.anpaglobal.org/conferences/{year}/abstract_booklet_full.pdf">
        
        <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/css/bootstrap.min.css">
        <link rel="icon" href="https://anpaglobal.org/wp-content/uploads/2021/05/favicon.ico" type="image/x-icon">
  </head>
           """
    
    return " ".join([str4, ''.join(str5), str6,
                     str1,''.join(str2), str3]
                     )
```

```python
def prep_list(df):
    Names_L= df['Author_I'].values
    title_L = df['presentation_title'].values
    affiliations_L = df['author_aff'].values
    abstract_L = df['abstract_text'].values
    abstract_n_L=df['abstract_id'].values
    Date_L=df['Date/Time'].astype(str).to_list()
    Date_L_N=df['Nepal Date/Time'].astype(str).to_list()
    Location_L = df['CDP/FIU/Remote'].values
    return Names_L,affiliations_L,title_L,abstract_L,abstract_n_L,Date_L,Date_L_N, Location_L
```

```python
prep_list(df_data)
```

```python
def create_html(df_name,html_name):
    Names_L,affiliations_L,title_L,abstract_L,abstract_n_L,Date_L,Date_L_N, Location_L =prep_list(df_name)
    fo = open(html_name, 'w',encoding="utf-8")
    fo.write(prepend)
    fo = open(html_name, 'a+',encoding="utf-8")
    for i in range(len(Names_L)):

        names_ = Names_L[i]
        title_ = title_L[i]
        affil_ = affiliations_L[i]
        abstract_ = abstract_L[i]
        abstract_n_=abstract_n_L[i]
        date_=Date_L[i]
        nepaldate_=Date_L_N[i]
        location_ = Location_L[i]

        html_code = '''

        <!-- ########################################################################## -->

        <tr>
                    <td width="158" colspan="2">
                        <p> Date/Time: <br>
                            ET:&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;{date_} </br>
                            Nepal: {nepaldate_}
                        </p>
                    </td>

        <td width="521" colspan="6">
            <p>Abstract Number: {abstract_n_} </p>

            <p>Presenting Author: {names_}
            </p>
            
            <p>Presenter's Affiliation: {affil_}
            </p>

            <p> <strong>Title: </strong>{title_}</p>

            <p> <strong>Location: </strong>{location_}</p>

            <p class="show_hide">Show/Hide Abstract</p>
            <div class="show_on_click">{abstract_}</div>

        </td>
        </tr>

    '''.format(date_=date_,nepaldate_=nepaldate_,abstract_n_=abstract_n_,affil_=affil_,names_=names_,
               title_ = title_,abstract_=abstract_,location_=location_)
        fo.write(html_code)

    fo = open(html_name, 'a+',encoding="utf-8")
    fo.write(postpend)
    fo.close()
```

```python
create_html(df_data,html_name='html_code_2023_data.html')
create_html(df_PER,html_name='html_code_2023_PER.html')
create_html(df_Condensed,html_name='html_code_2023_Condensed.html')
create_html(df_Bio,html_name='html_code_2023_Bio.html')
create_html(df_Plasma,html_name='html_code_2023_Plasma.html')
create_html(df_HEP,html_name='html_code_2023_HEP.html')
create_html(df_Applied,html_name='html_code_2023_Applied.html')
create_html(df_Astro,html_name='html_code_2023_Astro.html')

# Create HTML based on Location
create_html(df_CDP,html_name='html_code_2023_CDP.html')
create_html(df_FIU,html_name='html_code_2023_FIU.html')
```

```python
df['Start_DateTime']=df['Start Date'].astype(str)+' '+df['Start Time'].astype(str)
df['Start_DateTime'] = pd.to_datetime(df['Start_DateTime'],format=('%Y-%m-%d  %H:%M:%S'))
df['Start_DateTime'] = df['Start_DateTime'] + timedelta(hours=4)#.dt.tz_localize('UTC')#.dt.tz_convert('US/Eastern')

df['End_DateTime']=  df['End Date'].astype(str)+' '+df['End Time'].astype(str)
df['End_DateTime'] = pd.to_datetime(df['End_DateTime'],format=('%Y-%m-%d  %H:%M:%S'))
df['End_DateTime'] = df['End_DateTime'] + timedelta(hours=4)#.dt.tz_localize('UTC')#.dt.tz_convert('US/Eastern')#
```

```python
def prep_list(df):
    Names_L= df['Author_I'].values
    title_L = df['presentation_title'].values
    Date_L=df['Start_DateTime'].values
    EndDate_L=df['End_DateTime'].values
    return Names_L,title_L,Date_L,EndDate_L
```
